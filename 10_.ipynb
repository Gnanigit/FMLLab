```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
def remove_punc(sent):
  cal=""
  for i in sent:
    if i not in string.punctuation:
      cal=cal+i
  return cal
```

```python
def remove_sw(sent):
  stop_words=set(stopwords.words('english'))
  word_token=word_tokenize(sent)
  final=""
  for i in word_token:
    if i not in stop_words:
      final=final+i+" "
  return final
```

```python
sent="""We are learning* Natural !Language Processing (NLP) as part of
Fundamentals of Machine Learning (FML) course in our second year B.Tech."""
print("sentece after removing stop words : ",remove_sw(sent))
print("sentece after removing punctuations : ",remove_punc(sent))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,10,50)
plt.plot(x,tanh(x))
plt.xlabel("x terms")
plt.ylabel("tanh values")
plt.title("tanh activation function")
plt.show()
```
