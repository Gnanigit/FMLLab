```python
import numpy as np
import matplotlib.pyplot as plt
def tanh(x):
  return np.tanh(x)
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :tanh')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1 / (1+np.exp(-x))
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x), axis=0)
x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.show()
```
